## 1. –ü–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –∏ —Ç–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö (—Å–ª–∞–π–¥ 5)

–ü–µ—Ä–µ–º–µ–Ω–Ω–∞—è ‚Äî —ç—Ç–æ –∫–∞–∫ **–∫–æ—Ä–æ–±–∫–∞ —Å –Ω–∞–∫–ª–µ–π–∫–æ–π**:

- —É –∫–æ—Ä–æ–±–∫–∏ –µ—Å—Ç—å **–∏–º—è** (–∏–º—è –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π),
- –≤–Ω—É—Ç—Ä–∏ –ª–µ–∂–∏—Ç **–∑–Ω–∞—á–µ–Ω–∏–µ**,
- —É –∑–Ω–∞—á–µ–Ω–∏—è –µ—Å—Ç—å **—Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö**: —á–∏—Å–ª–æ, —Å—Ç—Ä–æ–∫–∞, –±—É–ª–µ–≤–æ –∏ —Ç.–¥.

–°–µ–π—á–∞—Å —Å–æ–∑–¥–∞–¥–∏–º –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –¥–ª—è –Ω–∞—à–µ–≥–æ –≥–µ—Ä–æ—è –§—É–Ω—Ç–∏–∫–∞ üê± –∏ –≤—ã–≤–µ–¥–µ–º –∏—Ö –Ω–∞ —ç–∫—Ä–∞–Ω.

```python
# üî¢ –ß–∏—Å–ª–æ–≤—ã–µ —Ç–∏–ø—ã
age = 25             # int ‚Äî —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ (–≤–æ–∑—Ä–∞—Å—Ç)
height = 1.75        # float ‚Äî –¥—Ä–æ–±–Ω–æ–µ —á–∏—Å–ª–æ (—Ä–æ—Å—Ç –≤ –º–µ—Ç—Ä–∞—Ö)

# üßµ –°—Ç—Ä–æ–∫–∞ (str)
name = "–§—É–Ω—Ç–∏–∫"      # –ò–º—è –∫–æ—Ç–∞-—Å—Ç—É–¥–µ–Ω—Ç–∞

# ‚úÖ –õ–æ–≥–∏—á–µ—Å–∫–∏–π —Ç–∏–ø (bool)
is_hungry = True     # –ì–æ–ª–æ–¥–µ–Ω –ª–∏ –∫–æ—Ç –ø—Ä—è–º–æ —Å–µ–π—á–∞—Å?

print("üë§ –ò–º—è:", name)
print("üéÇ –í–æ–∑—Ä–∞—Å—Ç:", age)
print("üìè –†–æ—Å—Ç:", height)
print("üòº –ì–æ–ª–æ–¥–µ–Ω –ª–∏ –§—É–Ω—Ç–∏–∫?", is_hungry)

# –ú–æ–∂–Ω–æ –∏–∑–º–µ–Ω–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π
age = age + 1
print("\nüéâ –ß–µ—Ä–µ–∑ –≥–æ–¥ –§—É–Ω—Ç–∏–∫—É –±—É–¥–µ—Ç:", age)

# –ü—Ä–æ–≤–µ—Ä–∏–º —Ç–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö
print("\nüß™ –¢–∏–ø –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π name:", type(name))
print("üß™ –¢–∏–ø –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π height:", type(height))
print("üß™ –¢–∏–ø –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π is_hungry:", type(is_hungry))
```

## 2. –°–ø–∏—Å–æ–∫ (List): –¥–µ–ª–∞ –§—É–Ω—Ç–∏–∫–∞ –Ω–∞ –Ω–µ–¥–µ–ª—é (—Å–ª–∞–π–¥ 6)

–°–ø–∏—Å–æ–∫ (`list`) ‚Äî —ç—Ç–æ **—É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω–∞—è –∫–æ–ª–ª–µ–∫—Ü–∏—è** –∑–Ω–∞—á–µ–Ω–∏–π.  
–ü—Ä–µ–¥—Å—Ç–∞–≤–∏–º —Å–ø–∏—Å–æ–∫ –¥–µ–ª –§—É–Ω—Ç–∏–∫–∞ –Ω–∞ –Ω–µ–¥–µ–ª—é:

- –∫–∞–∂–¥–æ–µ –¥–µ–ª–æ ‚Äî –æ—Ç–¥–µ–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç —Å–ø–∏—Å–∫–∞;
- –º—ã –º–æ–∂–µ–º:
  - –≤–∑—è—Ç—å —ç–ª–µ–º–µ–Ω—Ç –ø–æ –Ω–æ–º–µ—Ä—É (–∏–Ω–¥–µ–∫—Å—É),
  - –¥–æ–±–∞–≤–∏—Ç—å –Ω–æ–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç,
  - –ø—Ä–æ–π—Ç–∏—Å—å –ø–æ —Å–ø–∏—Å–∫—É –≤ —Ü–∏–∫–ª–µ.

–°–µ–π—á–∞—Å –ø–æ–≤—Ç–æ—Ä–∏–º –ø—Ä–∏–º–µ—Ä —Å–æ —Å–ª–∞–π–¥–∞ –∏ —á—É—Ç—å –µ–≥–æ —Ä–∞—Å—à–∏—Ä–∏–º.

```python
# üìã –°–ø–∏—Å–æ–∫ –¥–µ–ª –§—É–Ω—Ç–∏–∫–∞ –Ω–∞ –Ω–µ–¥–µ–ª—é

todo_list = [
    "üìö –ù–∞–ø–∏—Å–∞—Ç—å —ç—Å—Å–µ –ø–æ —ç–∫–æ–Ω–æ–º–∏–∫–µ",
    "üíª –í—ã–ø–æ–ª–Ω–∏—Ç—å –∑–∞–¥–∞–Ω–∏–µ –ø–æ Python",
    "üé¨ –ü–æ—Å–º–æ—Ç—Ä–µ—Ç—å –ª–µ–∫—Ü–∏—é –Ω–∞ YouTube",
    "‚òï –ö–æ—Ñ–µ —Å –æ–¥–Ω–æ–∫—É—Ä—Å–Ω–∏–∫–∞–º–∏",
    "üéÆ –ü–æ–∏–≥—Ä–∞—Ç—å –≤ –≤–∏–¥–µ–æ–∏–≥—Ä—ã",
]

print("üê± –î–µ–ª–∞ –§—É–Ω—Ç–∏–∫–∞:", todo_list)

# –î–æ—Å—Ç—É–ø –∫ —ç–ª–µ–º–µ–Ω—Ç–∞–º –ø–æ –∏–Ω–¥–µ–∫—Å—É
print("\n‚≠ê –°–∞–º–æ–µ –≤–∞–∂–Ω–æ–µ –¥–µ–ª–æ:", todo_list[0])     # –ø–µ—Ä–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç
print("üòå –°–∞–º–æ–µ –ø—Ä–∏—è—Ç–Ω–æ–µ –¥–µ–ª–æ:", todo_list[-1])   # –ø–æ—Å–ª–µ–¥–Ω–∏–π —ç–ª–µ–º–µ–Ω—Ç

# –î–æ–±–∞–≤–∏–º –Ω–æ–≤–æ–µ –¥–µ–ª–æ
todo_list.append("üèã –¢—Ä–µ–Ω–∏—Ä–æ–≤–∫–∞")
print("\n‚ûï –î–æ–±–∞–≤–∏–ª–∏ –Ω–æ–≤–æ–µ –¥–µ–ª–æ. –¢–µ–ø–µ—Ä—å —Å–ø–∏—Å–æ–∫ —Ç–∞–∫–æ–π:")
print(todo_list)

# –°–∫–æ–ª—å–∫–æ –≤—Å–µ–≥–æ –¥–µ–ª?
print(f"\nüìä –£ –§—É–Ω—Ç–∏–∫–∞ –µ—Å—Ç—å {len(todo_list)} –¥–µ–ª(–∞) –Ω–∞ –Ω–µ–¥–µ–ª—é")

# –ü—Ä–æ–π–¥—ë–º—Å—è –ø–æ —Å–ø–∏—Å–∫—É –≤ —Ü–∏–∫–ª–µ —Å –Ω—É–º–µ—Ä–∞—Ü–∏–µ–π
print("\nüìÖ –ü–æ–ª–Ω—ã–π —Å–ø–∏—Å–æ–∫ –¥–µ–ª –ø–æ –ø–æ—Ä—è–¥–∫—É:")
for i, task in enumerate(todo_list, start=1):
    print(f"  {i}. {task}")
```

## 3. –°–ø–∏—Å–æ–∫ –∫–∞–∫ —Å—Ç—Ä—É–∫—Ç—É—Ä–∞: –∏–Ω–¥–µ–∫—Å—ã –∏ –ø–æ—Ä—è–¥–æ–∫ (—Å–ª–∞–π–¥ 7)

–í–∞–∂–Ω–æ –ø–æ–º–Ω–∏—Ç—å:

- –∏–Ω–¥–µ–∫—Å—ã –≤ Python –Ω–∞—á–∏–Ω–∞—é—Ç—Å—è —Å **0**;
- –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ –∏–Ω–¥–µ–∫—Å—ã (`-1`, `-2`, ‚Ä¶) —Å—á–∏—Ç–∞—é—Ç —ç–ª–µ–º–µ–Ω—Ç—ã **—Å –∫–æ–Ω—Ü–∞** —Å–ø–∏—Å–∫–∞.

–ù–∞ —Å–ª–∞–π–¥–µ –±—ã–ª –ø—Ä–∏–º–µ—Ä —Å–æ —Å–ø–∏—Å–∫–æ–º —Ñ—Ä—É–∫—Ç–æ–≤. –ü–æ–≤—Ç–æ—Ä–∏–º –µ–≥–æ –≤ –∫–æ–¥–µ –∏ –ø–æ–∏–≥—Ä–∞–µ–º —Å –∏–Ω–¥–µ–∫—Å–∞–º–∏.

```python
# üçé –°–ø–∏—Å–æ–∫ —Ñ—Ä—É–∫—Ç–æ–≤

fruits = ["–Ø–±–ª–æ–∫–æ", "–ë–∞–Ω–∞–Ω", "–ê–ø–µ–ª—å—Å–∏–Ω", "–ö–∏–≤–∏"]

print("üß∫ –§—Ä—É–∫—Ç—ã:", fruits)

# –ò–Ω–¥–µ–∫—Å–∞—Ü–∏—è —Å –Ω–∞—á–∞–ª–∞
print("\n‚û° –ü–µ—Ä–≤—ã–π —Ñ—Ä—É–∫—Ç (–∏–Ω–¥–µ–∫—Å 0):", fruits[0])
print("‚û° –í—Ç–æ—Ä–æ–π —Ñ—Ä—É–∫—Ç (–∏–Ω–¥–µ–∫—Å 1):", fruits[1])

# –ò–Ω–¥–µ–∫—Å–∞—Ü–∏—è —Å –∫–æ–Ω—Ü–∞
print("\n‚¨Ö –ü–æ—Å–ª–µ–¥–Ω–∏–π —Ñ—Ä—É–∫—Ç (–∏–Ω–¥–µ–∫—Å -1):", fruits[-1])
print("‚¨Ö –ü—Ä–µ–¥–ø–æ—Å–ª–µ–¥–Ω–∏–π —Ñ—Ä—É–∫—Ç (–∏–Ω–¥–µ–∫—Å -2):", fruits[-2])

# –ú–æ–∂–Ω–æ –∑–∞–º–µ–Ω–∏—Ç—å —ç–ª–µ–º–µ–Ω—Ç –ø–æ –∏–Ω–¥–µ–∫—Å—É
fruits[1] = "–ú–∞–Ω–≥–æ"
print("\n‚úè –û–±–Ω–æ–≤–∏–ª–∏ –≤—Ç–æ—Ä–æ–π —ç–ª–µ–º–µ–Ω—Ç (–ë–∞–Ω–∞–Ω ‚Üí –ú–∞–Ω–≥–æ):")
print("üß∫ –¢–µ–ø–µ—Ä—å —Ñ—Ä—É–∫—Ç—ã:", fruits)
```

## 4. –¢–∏–ø—ã –∫–æ–ª–ª–µ–∫—Ü–∏–π: —Ç–∞–Ω—Ü—É—é—â–∏–π –∫–æ—Ç –∏ –∏–≥—Ä—É—à–∫–∏ ‚Äî —Ç–µ–æ—Ä–∏—è (—Å–ª–∞–π–¥ 8)

–í Python –µ—Å—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ –æ—Å–Ω–æ–≤–Ω—ã—Ö —Ç–∏–ø–æ–≤ –∫–æ–ª–ª–µ–∫—Ü–∏–π:

- **List (`list`)** ‚Äî –∏–∑–º–µ–Ω—è–µ–º—ã–π, —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ –∑–Ω–∞—á–µ–Ω–∏–π  
  –ü—Ä–∏–º–µ—Ä: –ø—Ä–æ–≥—Ä–∞–º–º–∞ –≤—ã—Å—Ç—É–ø–ª–µ–Ω–∏–π —Ç–∞–Ω—Ü—É—é—â–µ–≥–æ –∫–æ—Ç–∞.
- **Tuple (`tuple`)** ‚Äî –Ω–µ–∏–∑–º–µ–Ω—è–µ–º–∞—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å  
  –ü—Ä–∏–º–µ—Ä: –≤—Ä–µ–º—è –Ω–∞—á–∞–ª–∞ –∫–æ–Ω–∫—É—Ä—Å–∞ (—á–∞—Å—ã, –º–∏–Ω—É—Ç—ã).
- **Dict (`dict`)** ‚Äî –ø–∞—Ä—ã ¬´–∫–ª—é—á ‚Üí –∑–Ω–∞—á–µ–Ω–∏–µ¬ª  
  –ü—Ä–∏–º–µ—Ä: —Å–ª–æ–≤–∞—Ä—å —Å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–µ–π –æ —Ç–∞–Ω—Ü–∞—Ö –∏ —Ä–µ–∫–≤–∏–∑–∏—Ç–µ.
- **Set (`set`)** ‚Äî –º–Ω–æ–∂–µ—Å—Ç–≤–æ *—É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö* –∑–Ω–∞—á–µ–Ω–∏–π  
  –ü—Ä–∏–º–µ—Ä: –Ω–∞–±–æ—Ä –≤—Å–µ—Ö —Å—Ç–∏–ª–µ–π —Ç–∞–Ω—Ü–µ–≤ –±–µ–∑ –ø–æ–≤—Ç–æ—Ä–æ–≤.

–î–∞–ª—å—à–µ –º—ã —Å–æ–∑–¥–∞–¥–∏–º –ø–æ –æ–¥–Ω–æ–º—É –ø—Ä–∏–º–µ—Ä—É –∫–∞–∂–¥–æ–≥–æ —Ç–∏–ø–∞ –¥–ª—è –Ω–∞—à–µ–≥–æ –∫–æ—Ç–∞‚Äë—Ç–∞–Ω—Ü–æ—Ä–∞ üê±

```python
# üê± –¢–∞–Ω—Ü—É—é—â–∏–π –∫–æ—Ç: —á–µ—Ç—ã—Ä–µ —Ç–∏–ø–∞ –∫–æ–ª–ª–µ–∫—Ü–∏–π

# List ‚Äî —Ç–∞–Ω—Ü—ã –ø–æ –ø–æ—Ä—è–¥–∫—É
dance_skills = ["–ë–∞–ª—å–Ω—ã–π —Ç–∞–Ω–µ—Ü", "–ß–∞-—á–∞-—á–∞", "–†—ç–ø"]
print("üìã –°–ø–∏—Å–æ–∫ —Ç–∞–Ω—Ü–µ–≤ (list):", dance_skills)

# Tuple ‚Äî –≤—Ä–µ–º—è –Ω–∞—á–∞–ª–∞ –∫–æ–Ω–∫—É—Ä—Å–∞ (—á–∞—Å, –º–∏–Ω—É—Ç–∞)
start_time = (19, 30)
print("‚è∞ –í—Ä–µ–º—è –Ω–∞—á–∞–ª–∞ (tuple):", start_time)

# Dict ‚Äî —Ä–µ–∫–≤–∏–∑–∏—Ç –¥–ª—è —Ç–∞–Ω—Ü–µ–≤
props = {
    "–í–∞–ª—å—Å": "üß∂ –∫–ª—É–±–æ–∫",
    "–ë—Ä–µ–π–∫-–¥–∞–Ω—Å": "üéæ –º—è—á–∏–∫",
    "–†—ç–ø": "üê≠ –º—ã—à–∫–∞",
}
print("üìñ –†–µ–∫–≤–∏–∑–∏—Ç –¥–ª—è '–í–∞–ª—å—Å' (dict):", props["–í–∞–ª—å—Å"])

# Set ‚Äî —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ —Å—Ç–∏–ª–∏ (–¥—É–±–ª–∏–∫–∞—Ç—ã —É–±–∏—Ä–∞—é—Ç—Å—è)
all_styles = {"–í–∞–ª—å—Å", "–†—ç–ø", "–í–∞–ª—å—Å", "–•–∏–ø-—Ö–æ–ø"}
print("‚ú® –£–Ω–∏–∫–∞–ª—å–Ω—ã–µ —Å—Ç–∏–ª–∏ (set):", all_styles)
```

## 5. –¢–∏–ø—ã –∫–æ–ª–ª–µ–∫—Ü–∏–π: —Ç–∞–Ω—Ü—É—é—â–∏–π –∫–æ—Ç –∏ –∏–≥—Ä—É—à–∫–∏ ‚Äî –∫–æ–¥ (—Å–ª–∞–π–¥ 9)

–¢–µ–ø–µ—Ä—å –ø–æ–≤—Ç–æ—Ä–∏–º –ø–æ–ª–Ω—ã–π –ø—Ä–∏–º–µ—Ä —Å–æ —Å–ª–∞–π–¥–∞:

- —Å–ª–µ–≤–∞ ‚Äî `List + Tuple`: –ø—Ä–æ–≥—Ä–∞–º–º–∞ –≤—ã—Å—Ç—É–ø–ª–µ–Ω–∏–π –∏ –≤—Ä–µ–º—è –Ω–∞—á–∞–ª–∞;
- —Å–ø—Ä–∞–≤–∞ ‚Äî `Dict + Set`: —Ä–µ–∫–≤–∏–∑–∏—Ç –¥–ª—è –Ω–æ–º–µ—Ä–∞ –∏ –º–Ω–æ–∂–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å—Ç–∏–ª–µ–π.

–í –Ω–æ—É—Ç–±—É–∫–µ –æ–±—ä–µ–¥–∏–Ω–∏–º –≤—Å—ë –≤ –æ–¥–∏–Ω –±–ª–æ–∫ –∫–æ–¥–∞, —á—Ç–æ–±—ã –≤–∏–¥–µ—Ç—å –æ–±—â—É—é –∫–∞—Ä—Ç–∏–Ω—É.

```python
# üê± –§—É–Ω—Ç–∏–∫ –Ω–∞ –∫–æ–Ω–∫—É—Ä—Å–µ —Ç–∞–Ω—Ü–µ–≤

# List ‚Äî —Ç–∞–Ω—Ü—ã –ø–æ –ø–æ—Ä—è–¥–∫—É –≤—ã—Å—Ç—É–ø–ª–µ–Ω–∏—è
performance = ["–ë–∞–ª—å–Ω—ã–π —Ç–∞–Ω–µ—Ü", "–ß–∞-—á–∞-—á–∞", "–†—ç–ø"]
performance.append("–ë—Ä–µ–π–∫-–¥–∞–Ω—Å")  # –¥–æ–±–∞–≤–∏–ª–∏ —Å—é—Ä–ø—Ä–∏–∑

# Tuple ‚Äî –≤—Ä–µ–º—è –Ω–∞—á–∞–ª–∞ –∫–æ–Ω–∫—É—Ä—Å–∞ (—á–∞—Å—ã, –º–∏–Ω—É—Ç—ã)
start_time = (19, 30)

print("üé≠ –ü—Ä–æ–≥—Ä–∞–º–º–∞ –≤—ã—Å—Ç—É–ø–ª–µ–Ω–∏–π:")
for i, dance in enumerate(performance, start=1):
    print(f"  {i}. {dance}")

print(f"\n‚è∞ –ù–∞—á–∞–ª–æ –∫–æ–Ω–∫—É—Ä—Å–∞ –≤ {start_time[0]}:{start_time[1]:02d}")

# Dict ‚Äî —Ä–µ–∫–≤–∏–∑–∏—Ç –¥–ª—è –Ω–æ–º–µ—Ä–∞
props = {
    "–í–∞–ª—å—Å": "üß∂ –∫–ª—É–±–æ–∫",
    "–ë—Ä–µ–π–∫-–¥–∞–Ω—Å": "üéæ –º—è—á–∏–∫",
    "–†—ç–ø": "üê≠ –º—ã—à–∫–∞",
}

# Set ‚Äî —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ —Å—Ç–∏–ª–∏ (–µ—Å–ª–∏ –∫—Ç–æ-—Ç–æ –ø–æ–≤—Ç–æ—Ä–∏–ª—Å—è ‚Äî –æ—Å—Ç–∞–Ω–µ—Ç—Å—è –æ–¥–∏–Ω —Ä–∞–∑)
all_styles = {"–í–∞–ª—å—Å", "–†—ç–ø", "–í–∞–ª—å—Å", "–•–∏–ø-—Ö–æ–ø"}

print("\nüéÅ –†–µ–∫–≤–∏–∑–∏—Ç –¥–ª—è –±—Ä–µ–π–∫-–¥–∞–Ω—Å–∞:", props["–ë—Ä–µ–π–∫-–¥–∞–Ω—Å"])
print("‚ú® –£–Ω–∏–∫–∞–ª—å–Ω—ã–µ —Å—Ç–∏–ª–∏ —Ç–∞–Ω—Ü–µ–≤:", all_styles)
```

## 6. –°–ª–æ–≤–∞—Ä–∏ (dict): –ø—Ä–æ—Ñ–∏–ª–∏ –∫–æ—Ç–æ–≤ –∏ –∏–≥—Ä—É—à–∫–∏ (—Å–ª–∞–π–¥ 10)

–°–ª–æ–≤–∞—Ä—å (`dict`) —Ö—Ä–∞–Ω–∏—Ç –ø–∞—Ä—ã ¬´–∫–ª—é—á ‚Üí –∑–Ω–∞—á–µ–Ω–∏–µ¬ª.  
–≠—Ç–æ —É–¥–æ–±–Ω–æ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è **–ø—Ä–æ—Ñ–∏–ª–µ–π –æ–±—ä–µ–∫—Ç–æ–≤**:

- –∫–ª—é—á–∏ ‚Äî –∏–º–µ–Ω–∞ –ø–æ–ª–µ–π (`"name"`, `"university"`, `"toys"`),
- –∑–Ω–∞—á–µ–Ω–∏—è ‚Äî –∫–æ–Ω–∫—Ä–µ—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ (—Å—Ç—Ä–æ–∫–∞, —á–∏—Å–ª–æ, —Å–ø–∏—Å–æ–∫ –∏ —Ç.–¥.).

–°–µ–π—á–∞—Å —Å–æ–∑–¥–∞–¥–∏–º –¥–≤–∞ –ø—Ä–æ—Ñ–∏–ª—è:

- –ë–∞–π—Ç–∏–∫ ‚Äî –∫–æ—Ç‚Äë—Å—Ç—É–¥–µ–Ω—Ç –∏–∑ MIT;
- –§—É–Ω—Ç–∏–∫ ‚Äî –∫–æ—Ç‚Äë—Å—Ç—É–¥–µ–Ω—Ç –∏–∑ –ü–µ—Ç—Ä–ì–£.

–ò –≤—ã–≤–µ–¥–µ–º –∏—Ö –¥–∞–Ω–Ω—ã–µ –≤ —á–∏—Ç–∞–±–µ–ª—å–Ω–æ–º –≤–∏–¥–µ.

```python
# üê± –ü—Ä–æ—Ñ–∏–ª—å –ë–∞–π—Ç–∏–∫–∞

bytik = {
    "name": "–ë–∞–π—Ç–∏–∫",
    "university": "MIT",
    "projects": 5,
    "toys": ["üê≠ –º—ã—à–∫–∞", "üéæ –º—è—á–∏–∫"],
}

print(f"üíª {bytik['name']} —É—á–∏—Ç—Å—è –≤ {bytik['university']}")
print(f"üìä –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–µ–∫—Ç–æ–≤: {bytik['projects']}")
print(f"üéÅ –ò–≥—Ä—É—à–∫–∏ –ë–∞–π—Ç–∏–∫–∞: {', '.join(bytik['toys'])}")

# üê± –ü—Ä–æ—Ñ–∏–ª—å –§—É–Ω—Ç–∏–∫–∞

funtik = {
    "name": "–§—É–Ω—Ç–∏–∫",
    "university": "–ü–µ—Ç—Ä–ì–£",
    "toys": ["üß∂ –∫–ª—É–±–æ–∫", "üêü —Ä—ã–±–∫–∞"],
}

print("\nüìö –ü—Ä–æ—Ñ–∏–ª—å –§—É–Ω—Ç–∏–∫–∞:")
print(f"üìö {funtik['name']} —É—á–∏—Ç—Å—è –≤ {funtik['university']}")
print(f"üéØ –ò–≥—Ä—É—à–∫–∏ –§—É–Ω—Ç–∏–∫–∞: {', '.join(funtik['toys'])}")

# –ù–µ–±–æ–ª—å—à–æ–µ —É–ø—Ä–∞–∂–Ω–µ–Ω–∏–µ: –¥–æ–±–∞–≤–∏–º –Ω–æ–≤—É—é –∏–≥—Ä—É—à–∫—É –§—É–Ω—Ç–∏–∫—É
funtik["toys"].append("üéæ –º—è—á–∏–∫")
print("\n‚ûï –î–æ–±–∞–≤–∏–ª–∏ –§—É–Ω—Ç–∏–∫—É –Ω–æ–≤—É—é –∏–≥—Ä—É—à–∫—É:")
print(f"üéØ –û–±–Ω–æ–≤–ª—ë–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ –∏–≥—Ä—É—à–µ–∫ –§—É–Ω—Ç–∏–∫–∞: {', '.join(funtik['toys'])}")
```
