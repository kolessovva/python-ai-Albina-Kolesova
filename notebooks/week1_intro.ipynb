## ‚ö° –ü–µ—Ä–≤—ã–π –ø—Ä–∏–º–µ—Ä: –ª–æ–≥–∏—á–µ—Å–∫–∏–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ

**–õ–æ–≥–∏—á–µ—Å–∫–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è (bool)** –º–æ–∂–µ—Ç –∏–º–µ—Ç—å —Ç–æ–ª—å–∫–æ –¥–≤–∞ –∑–Ω–∞—á–µ–Ω–∏—è: `True` –∏–ª–∏ `False`.

–ü–æ–ø—Ä–æ–±—É–π—Ç–µ –∏–∑–º–µ–Ω–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ `E` –Ω–∞ `False` –∏ –ø–æ—Å–º–æ—Ç—Ä–∏—Ç–µ, —á—Ç–æ –ø—Ä–æ–∏–∑–æ–π–¥—ë—Ç!

**–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ** –Ω–∞ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ `and` –≤–æ –≤—Ç–æ—Ä–æ–º –ø—Ä–∏–º–µ—Ä–µ ‚Äî –æ–Ω –ø—Ä–æ–≤–µ—Ä—è–µ—Ç –û–ë–ê —É—Å–ª–æ–≤–∏—è –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ.

````python
# –ü—Ä–æ—Å—Ç–æ–π –ø—Ä–∏–º–µ—Ä
E = True  # –≠–Ω–µ—Ä–≥–∏—è
if E:
    print("‚ö° –ß–∞–π–Ω–∏–∫ –≤–∫–ª—é—á—ë–Ω!")
else:
    print("‚ùå –ß–∞–π–Ω–∏–∫ –≤—ã–∫–ª—é—á–µ–Ω")

print()  # –ü—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞ –¥–ª—è —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è

# –°–ª–æ–∂–Ω—ã–π –ø—Ä–∏–º–µ—Ä
E = True   # –ï—Å—Ç—å —ç–Ω–µ—Ä–≥–∏—è
W = True   # –ï—Å—Ç—å –≤–æ–¥–∞

if E and W:
    print("‚úÖ –ß–∞–π–Ω–∏–∫ –≤–∫–ª—é—á—ë–Ω –∏ –≤–æ–¥–∞ –Ω–∞–≥—Ä–µ–≤–∞–µ—Ç—Å—è!")
elif E and not W:
    print("‚ö†Ô∏è –ß–∞–π–Ω–∏–∫ –≤–∫–ª—é—á—ë–Ω, –Ω–æ –≤–æ–¥—ã –Ω–µ—Ç!")
elif not E and W:
    print("‚ö†Ô∏è –í–æ–¥–∞ –µ—Å—Ç—å, –Ω–æ –Ω–µ—Ç —ç–ª–µ–∫—Ç—Ä–∏—á–µ—Å—Ç–≤–∞!")
else:
    print("‚ùå –ù–µ—Ç –Ω–∏ –≤–æ–¥—ã, –Ω–∏ —ç–ª–µ–∫—Ç—Ä–∏—á–µ—Å—Ç–≤–∞!")
````

## üë• –†–∞–±–æ—Ç–∞ —Å–æ —Å–ø–∏—Å–∫–∞–º–∏: —Å–∫–∞–∑–∫–∞ –ø—Ä–æ —Ä–µ–ø–∫—É

–ó–¥–µ—Å—å –º—ã –≤–∏–¥–∏–º:
- –°–æ–∑–¥–∞–Ω–∏–µ —Å–ø–∏—Å–∫–∞ —Å —ç–º–æ–¥–∑–∏
- –°—Ä–µ–∑ —Å–ø–∏—Å–∫–∞ `heroes[1:]` ‚Äî –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã, –∫—Ä–æ–º–µ –ø–µ—Ä–≤–æ–≥–æ
- –¶–∏–∫–ª `for` –¥–ª—è –ø–µ—Ä–µ–±–æ—Ä–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤
- –ö–æ–Ω–∫–∞—Ç–µ–Ω–∞—Ü–∏—é —Å—Ç—Ä–æ–∫ —á–µ—Ä–µ–∑ `+=`

**–ü–æ–ø—Ä–æ–±—É–π—Ç–µ:** –¥–æ–±–∞–≤—å—Ç–µ –≤ —Å–ø–∏—Å–æ–∫ –Ω–æ–≤–æ–≥–æ –≥–µ—Ä–æ—è –∏–ª–∏ –∏–∑–º–µ–Ω–∏—Ç–µ –ø–æ—Ä—è–¥–æ–∫!

````python
# –ì–µ—Ä–æ–∏ —Å–∫–∞–∑–∫–∏ –ø—Ä–æ —Ä–µ–ø–∫—É
heroes = [
    "üëµ –ë–∞–±–∫–∞",
    "üë¥ –î–µ–¥–∫–∞",
    "üë©‚Äçüåæ –í–Ω—É—á–∫–∞",
    "üêï –ñ—É—á–∫–∞",
    "üê± –ö–æ—à–∫–∞",
    "üê≠ –ú—ã—à–∫–∞"
]

# –ù–∞—á–∏–Ω–∞–µ–º —Å –±–∞–±–∫–∏
chain = heroes[0]

# –ü–æ—Å—Ç–µ–ø–µ–Ω–Ω–æ –¥–æ–±–∞–≤–ª—è–µ–º –≥–µ—Ä–æ–µ–≤
for hero in heroes[1:]:
    chain += " ‚Üí " + hero

print("üåæ –ö—Ç–æ —Ç—è–Ω—É–ª —Ä–µ–ø–∫—É:")
print(chain)
print(f"\nüìä –í—Å–µ–≥–æ –≥–µ—Ä–æ–µ–≤: {len(heroes)}")
````

## üê± –ü–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –∏ —Ç–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö

–í Python –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è ‚Äî —ç—Ç–æ ¬´–∫–æ—Ä–æ–±–∫–∞¬ª –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö.

**–û—Å–Ω–æ–≤–Ω—ã–µ —Ç–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö:**
- `int` ‚Äî —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞ (–≤–æ–∑—Ä–∞—Å—Ç, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ)
- `float` ‚Äî –¥—Ä–æ–±–Ω—ã–µ —á–∏—Å–ª–∞ (–≤–µ—Å, —Ä–æ—Å—Ç)
- `str` ‚Äî —Ç–µ–∫—Å—Ç (–∏–º–µ–Ω–∞, –æ–ø–∏—Å–∞–Ω–∏—è)
- `bool` ‚Äî –ª–æ–≥–∏—á–µ—Å–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è (`True` –∏–ª–∏ `False`)

**–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ:**
- –ü–µ—Ä–µ–º–µ–Ω–Ω–æ–π –ø—Ä–æ—Å—Ç–æ –ø—Ä–∏—Å–≤–∞–∏–≤–∞–µ—Ç—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ —á–µ—Ä–µ–∑ `=`
- Python –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö
- –ò—Å–ø–æ–ª—å–∑—É–µ–º `f-—Å—Ç—Ä–æ–∫–∏` –¥–ª—è –≤—Å—Ç–∞–≤–∫–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –≤ —Ç–µ–∫—Å—Ç: `f"–¢–µ–∫—Å—Ç {–ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è}"`

````python
def get_age_suffix(age):
    if age % 100 in [11, 12, 13, 14]:
        return "–ª–µ—Ç"
    if age % 10 == 1:
        return "–≥–æ–¥"
    elif age % 10 in [2, 3, 4]:
        return "–≥–æ–¥–∞"
    else:
        return "–ª–µ—Ç"

# –ö–æ—Ç –§—É–Ω—Ç–∏–∫ (—ç–∫–æ–Ω–æ–º–∏—Å—Ç)
name = "–§—É–Ω—Ç–∏–∫"        # str - —Ç–µ–∫—Å—Ç üìù
age = 11                # int - —Ü–µ–ª–æ–µ üî¢
weight = 4.5           # float - –¥—Ä–æ–±–Ω–æ–µ ‚öñÔ∏è
is_hungry = True       # bool - –¥–∞/–Ω–µ—Ç üçΩÔ∏è

suffix = get_age_suffix(age)
print(f"üê± {name}, –≤–æ–∑—Ä–∞—Å—Ç {age} {suffix}")
print(f"–í–µ—Å: {weight} –∫–≥")
print(f"–ì–æ–ª–æ–¥–µ–Ω? {is_hungry}")
````

## üßÆ –ê—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–∏

Python —É–º–µ–µ—Ç –≤—ã–ø–æ–ª–Ω—è—Ç—å –≤—Å–µ –±–∞–∑–æ–≤—ã–µ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–∏.

**–ë–∞–∑–æ–≤—ã–µ –æ–ø–µ—Ä–∞—Ü–∏–∏:**
- `+` ‚Äî —Å–ª–æ–∂–µ–Ω–∏–µ
- `-` ‚Äî –≤—ã—á–∏—Ç–∞–Ω–∏–µ
- `*` ‚Äî —É–º–Ω–æ–∂–µ–Ω–∏–µ
- `/` ‚Äî –¥–µ–ª–µ–Ω–∏–µ (–≤—Å–µ–≥–¥–∞ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –¥—Ä–æ–±–Ω–æ–µ —á–∏—Å–ª–æ)

**–°–ø–µ—Ü–∏–∞–ª—å–Ω—ã–µ –æ–ø–µ—Ä–∞—Ü–∏–∏:**
- `//` ‚Äî **—Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω–æ–µ –¥–µ–ª–µ–Ω–∏–µ** (—Ç–æ–ª—å–∫–æ —Ü–µ–ª–∞—è —á–∞—Å—Ç—å)
- `%` ‚Äî **–æ—Å—Ç–∞—Ç–æ–∫ –æ—Ç –¥–µ–ª–µ–Ω–∏—è**
- `**` ‚Äî **–≤–æ–∑–≤–µ–¥–µ–Ω–∏–µ –≤ —Å—Ç–µ–ø–µ–Ω—å**

**–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ:**
- –í –ø—Ä–∏–º–µ—Ä–µ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ `{per_day:.1f}` ‚Äî –≤—ã–≤–æ–¥–∏—Ç —á–∏—Å–ª–æ —Å –æ–¥–Ω–∏–º –∑–Ω–∞–∫–æ–º –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π

````python
# –§—É–Ω—Ç–∏–∫ —Å—á–∏—Ç–∞–µ—Ç —Å–≤–æ–∏ —Ä–∞—Å—Ö–æ–¥—ã üí∞

# –†–∞—Å—Ö–æ–¥—ã –∑–∞ –Ω–µ–¥–µ–ª—é
food = 100      # —Ñ—É–Ω—Ç–æ–≤ –Ω–∞ –µ–¥—É
toys = 50       # —Ñ—É–Ω—Ç–æ–≤ –Ω–∞ –∏–≥—Ä—É—à–∫–∏
vet = 200       # —Ñ—É–Ω—Ç–æ–≤ –Ω–∞ –≤–µ—Ç–µ—Ä–∏–Ω–∞—Ä–∞

# –û–±—â–∏–µ —Ä–∞—Å—Ö–æ–¥—ã
total = food + toys + vet
print(f"üí∞ –í—Å–µ–≥–æ —Ä–∞—Å—Ö–æ–¥–æ–≤: {total} —Ñ—É–Ω—Ç–æ–≤")

# –î–µ–ª–∏–º –Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π (7)
per_day = total / 7
print(f"üìä –í –¥–µ–Ω—å: {per_day:.1f} —Ñ—É–Ω—Ç–æ–≤")

# –ï—Å–ª–∏ –§—É–Ω—Ç–∏–∫ —Å—ä–µ–¥–∞–µ—Ç 50 –≥ –≤ –¥–µ–Ω—å
grams_per_day = 50
days = 7
total_grams = grams_per_day * days
print(f"üçΩÔ∏è –ó–∞ –Ω–µ–¥–µ–ª—é: {total_grams} –≥—Ä–∞–º–º–æ–≤ –∫–æ—Ä–º–∞")
````

## üîÄ –£—Å–ª–æ–≤–Ω—ã–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã (if, elif, else)

–£—Å–ª–æ–≤–Ω—ã–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã –ø–æ–∑–≤–æ–ª—è—é—Ç –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø—Ä–∏–Ω–∏–º–∞—Ç—å —Ä–µ—à–µ–Ω–∏—è.

**–°—Ç—Ä—É–∫—Ç—É—Ä–∞:**
```python
if —É—Å–ª–æ–≤–∏–µ:
    # –≤—ã–ø–æ–ª–Ω–∏—Ç—Å—è, –µ—Å–ª–∏ —É—Å–ª–æ–≤–∏–µ –∏—Å—Ç–∏–Ω–Ω–æ
elif –¥—Ä—É–≥–æ–µ_—É—Å–ª–æ–≤–∏–µ:
    # –≤—ã–ø–æ–ª–Ω–∏—Ç—Å—è, –µ—Å–ª–∏ –ø–µ—Ä–≤–æ–µ –ª–æ–∂–Ω–æ, –∞ —ç—Ç–æ –∏—Å—Ç–∏–Ω–Ω–æ
else:
    # –≤—ã–ø–æ–ª–Ω–∏—Ç—Å—è, –µ—Å–ª–∏ –≤—Å–µ —É—Å–ª–æ–≤–∏—è –ª–æ–∂–Ω—ã
```

**–û–ø–µ—Ä–∞—Ç–æ—Ä—ã —Å—Ä–∞–≤–Ω–µ–Ω–∏—è:**
- `==` ‚Äî —Ä–∞–≤–Ω–æ
- `!=` ‚Äî –Ω–µ —Ä–∞–≤–Ω–æ
- `<` ‚Äî –º–µ–Ω—å—à–µ
- `>` ‚Äî –±–æ–ª—å—à–µ
- `<=` ‚Äî –º–µ–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–æ
- `>=` ‚Äî –±–æ–ª—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–æ

**‚ö†Ô∏è –í–∞–∂–Ω–æ:** –ù–µ –ø—É—Ç–∞–π—Ç–µ `=` (–ø—Ä–∏—Å–≤–∞–∏–≤–∞–Ω–∏–µ) –∏ `==` (—Å—Ä–∞–≤–Ω–µ–Ω–∏–µ)!

````python
# –í–æ–∑—Ä–∞—Å—Ç –§—É–Ω—Ç–∏–∫–∞ üê±
age = 2

if age < 1:
    print("üê± –ö–æ—Ç—ë–Ω–æ–∫!")
elif age < 3:
    print("üê± –ú–æ–ª–æ–¥–æ–π –∫–æ—Ç!")
elif age < 10:
    print("üê± –í–∑—Ä–æ—Å–ª—ã–π –∫–æ—Ç!")
else:
    print("üê± –ú—É–¥—Ä—ã–π –∫–æ—Ç!")

# –†–µ–∑—É–ª—å—Ç–∞—Ç: üê± –ú–æ–ª–æ–¥–æ–π –∫–æ—Ç!
````

## üîó –õ–æ–≥–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã (and, or, not)

–õ–æ–≥–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã –ø–æ–∑–≤–æ–ª—è—é—Ç –∫–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ —É—Å–ª–æ–≤–∏–π.

**–¢—Ä–∏ –æ—Å–Ω–æ–≤–Ω—ã—Ö –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞:**
- `and` ‚Äî **–ò** (–æ–±–∞ —É—Å–ª–æ–≤–∏—è –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å True)
- `or` ‚Äî **–ò–õ–ò** (—Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ —É—Å–ª–æ–≤–∏–µ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å True)
- `not` ‚Äî **–ù–ï** (–ø–µ—Ä–µ–≤–æ—Ä–∞—á–∏–≤–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ)

**–ü—Ä–∏–º–µ—Ä—ã:**
- `age > 1 and age < 5` ‚Äî –≤–æ–∑—Ä–∞—Å—Ç –æ—Ç 2 –¥–æ 4
- `weight < 3 or weight > 7` ‚Äî –≤–µ—Å –ª–∏–±–æ –º–µ–Ω—å—à–µ 3, –ª–∏–±–æ –±–æ–ª—å—à–µ 7
- `not is_hungry` ‚Äî –ù–ï –≥–æ–ª–æ–¥–µ–Ω

**–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ:** –≤ –ø—Ä–∏–º–µ—Ä–µ –æ–±–∞ –∫–æ—Ç–∞ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—Ç —É—Å–ª–æ–≤–∏—è–º (–æ–±–∞ –∑–Ω–∞—á–µ–Ω–∏—è True), –ø–æ—ç—Ç–æ–º—É –ø–æ–ø–∞–¥–∞—é—Ç –≤ —Å–≤–æ–∏ ¬´—Ñ–∞–∫—É–ª—å—Ç–µ—Ç—ã¬ª.

````python
# üé© –ú–∞–≥–∏—á–µ—Å–∫–∞—è —à–ª—è–ø–∞ –¥–ª—è –∫–æ—Ç–æ–≤

# –ö–æ—Ç –ë–∞–π—Ç–∏–∫ üê±üíª
print("üé© –¢–µ—Å—Ç–∏—Ä—É–µ–º –ë–∞–π—Ç–∏–∫–∞...")
loves_code = True
works_at_night = False

if loves_code and (not (not works_at_night)):
    print("ü¶Å –ì–†–ò–§–§–ò–ù–î–û–†!")
    print("   –ë–∞–π—Ç–∏–∫ ‚Äî —Å–º–µ–ª—ã–π –ø—Ä–æ–≥—Ä–∞–º–º–∏—Å—Ç! üíª")
    print("   üöÄ –°–æ–∑–¥–∞—ë—Ç –∏–Ω–Ω–æ–≤–∞—Ü–∏–æ–Ω–Ω—ã–µ –ø—Ä–æ–µ–∫—Ç—ã!")
else:
    print("üìö –®–ª—è–ø–∞ –¥—É–º–∞–µ—Ç...")

# –ö–æ—Ç –§—É–Ω—Ç–∏–∫ üê±üìä
print("\nüé© –¢–µ—Å—Ç–∏—Ä—É–µ–º –§—É–Ω—Ç–∏–∫–∞...")
analyzes_markets = True
makes_forecasts = True

if analyzes_markets and makes_forecasts:
    print("ü¶Ö –ö–û–ì–¢–ï–í–†–ê–ù-–≠–ö–û–ù–û–ú–ò–ö–£–°!")
    print("   –§—É–Ω—Ç–∏–∫ ‚Äî –º—É–¥—Ä—ã–π —ç–∫–æ–Ω–æ–º–∏—Å—Ç! üìä")
    print("   üìà –í–∏–¥–∏—Ç –±—É–¥—É—â–µ–µ –≤ –¥–∞–Ω–Ω—ã—Ö!")
else:
    print("üìö –®–ª—è–ø–∞ –¥—É–º–∞–µ—Ç...")
````

## üîÅ –¶–∏–∫–ª for

–¶–∏–∫–ª `for` –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –¥–ª—è –ø–µ—Ä–µ–±–æ—Ä–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ (—Å–ø–∏—Å–∫–∞, —Å—Ç—Ä–æ–∫–∏ –∏ —Ç.–¥.).

**–°—Ç—Ä—É–∫—Ç—É—Ä–∞:**
```python
for —ç–ª–µ–º–µ–Ω—Ç in –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å:
    # –¥–µ–π—Å—Ç–≤–∏—è —Å —ç–ª–µ–º–µ–Ω—Ç–æ–º
```

**–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ:**
- `days` ‚Äî —ç—Ç–æ —Å–ø–∏—Å–æ–∫ –¥–Ω–µ–π –Ω–µ–¥–µ–ª–∏
- –¶–∏–∫–ª `for` –ø—Ä–æ—Ö–æ–¥–∏—Ç –ø–æ –∫–∞–∂–¥–æ–º—É —ç–ª–µ–º–µ–Ω—Ç—É —Å–ø–∏—Å–∫–∞
- –í–Ω—É—Ç—Ä–∏ —Ü–∏–∫–ª–∞ –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —É—Å–ª–æ–≤–∏—è `if`
- –°–∏–º–≤–æ–ª `\n` –¥–æ–±–∞–≤–ª—è–µ—Ç –Ω–æ–≤—É—é —Å—Ç—Ä–æ–∫—É –¥–ª—è –ª—É—á—à–µ–π —á–∏—Ç–∞–µ–º–æ—Å—Ç–∏

````python
# –§—É–Ω—Ç–∏–∫ —É—á–∏—Ç—Å—è –≤ —É–Ω–∏–≤–µ—Ä—Å–∏—Ç–µ—Ç–µ üéì

days = ["–ü–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫", "–í—Ç–æ—Ä–Ω–∏–∫", "–°—Ä–µ–¥–∞",
        "–ß–µ—Ç–≤–µ—Ä–≥–∏—É—Å", "–ß–µ—Ç–≤–µ—Ä–≥", "–ü—è—Ç–Ω–∏—Ü–∞", "–°—É–±–±–æ—Ç–∞", "–í–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ"]

for day in days:
    if day in ["–°—É–±–±–æ—Ç–∞", "–í–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ", "–ß–µ—Ç–≤–µ—Ä–≥–∏—É—Å"]:
        print(f"üê± {day}: –§—É–Ω—Ç–∏–∫ –æ—Ç–¥—ã—Ö–∞–µ—Ç! üò¥")
        print(f"    –°–ø–∏—Ç, –ª–µ–Ω–∏—Ç—Å—è, –∑–∞–≤–∏—Å–∞–µ—Ç –≤ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç–µ")
    else:
        print(f"üê± {day}: –§—É–Ω—Ç–∏–∫ –Ω–∞ –ø–∞—Ä–µ! üìö")
        print(f"    –£—á–∏—Ç Data Science!")

print("\nüìä –ù–µ–¥–µ–ª—è –ø—Ä–æ–π–¥–µ–Ω–∞!")
````

## ‚è±Ô∏è –¶–∏–∫–ª while

–¶–∏–∫–ª `while` –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è, –ø–æ–∫–∞ —É—Å–ª–æ–≤–∏–µ –∏—Å—Ç–∏–Ω–Ω–æ.

**–°—Ç—Ä—É–∫—Ç—É—Ä–∞:**
```python
while —É—Å–ª–æ–≤–∏–µ:
    # –¥–µ–π—Å—Ç–≤–∏—è
    # –Ω–µ –∑–∞–±—É–¥—å—Ç–µ –∏–∑–º–µ–Ω–∏—Ç—å —É—Å–ª–æ–≤–∏–µ, –∏–Ω–∞—á–µ –±—É–¥–µ—Ç –±–µ—Å–∫–æ–Ω–µ—á–Ω—ã–π —Ü–∏–∫–ª!
```

**–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ:**
- –°—á—ë—Ç—á–∏–∫ `attempts` —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è –Ω–∞ –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏
- –ö–æ–≥–¥–∞ `attempts` —Å—Ç–∞–Ω–µ—Ç —Ä–∞–≤–Ω—ã–º 3, —É—Å–ª–æ–≤–∏–µ `attempts < 3` —Å—Ç–∞–Ω–µ—Ç –ª–æ–∂–Ω—ã–º
- –¶–∏–∫–ª –æ—Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è

**–ë–æ–Ω—É—Å:** –≤—Ç–æ—Ä–æ–π –ø—Ä–∏–º–µ—Ä –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –æ–±—Ä–∞—Ç–Ω—ã–π –æ—Ç—Å—á—ë—Ç —Å –ø–æ–º–æ—â—å—é `range(5, 0, -1)`

````python
# –ë–∞–π—Ç–∏–∫ –ø—Ä–æ–≤–µ—Ä—è–µ—Ç –ø—Ä–æ–≥—Ä–∞–º–º—É üíª

attempts = 0

while attempts < 3:
    print(f"üê± –ü–æ–ø—ã—Ç–∫–∞ {attempts + 1}")
    print(f"   –ó–∞–ø—É—Å–∫ –ø—Ä–æ–≥—Ä–∞–º–º—ã...")
    attempts = attempts + 1

print("‚úÖ –ü—Ä–æ–≥—Ä–∞–º–º–∞ –∑–∞–ø—É—Å—Ç–∏–ª–∞—Å—å!")

print("\n" + "="*30)
````

````python
# –ó–∞–ø—É—Å–∫ —Ä–∞–∫–µ—Ç—ã! üöÄ
print("üöÄ –ó–ê–ü–£–°–ö!")
print("="*30)

for i in range(5, 0, -1):
    print(f"‚è±Ô∏è  {i}...")

print("="*30)
print("üöÄ –í–ó–õ–Å–¢!!! üåå")
````

## üìã –°–ø–∏—Å–∫–∏ (List)

–°–ø–∏—Å–æ–∫ ‚Äî —ç—Ç–æ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω–∞—è –∫–æ–ª–ª–µ–∫—Ü–∏—è –∑–Ω–∞—á–µ–Ω–∏–π.

**–í–∞–∂–Ω—ã–µ –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏:**
- –ò–Ω–¥–µ–∫—Å—ã –Ω–∞—á–∏–Ω–∞—é—Ç—Å—è —Å **0** (–ø–µ—Ä–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç ‚Äî `list[0]`)
- –ú–æ–∂–Ω–æ –æ–±—Ä–∞—â–∞—Ç—å—Å—è —Å –∫–æ–Ω—Ü–∞: `list[-1]` ‚Äî –ø–æ—Å–ª–µ–¥–Ω–∏–π —ç–ª–µ–º–µ–Ω—Ç
- `append()` ‚Äî –¥–æ–±–∞–≤–∏—Ç—å —ç–ª–µ–º–µ–Ω—Ç –≤ –∫–æ–Ω–µ—Ü —Å–ø–∏—Å–∫–∞
- `len()` ‚Äî –ø–æ–ª—É—á–∏—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤
- `enumerate()` ‚Äî –ø–æ–ª—É—á–∏—Ç—å –∏–Ω–¥–µ–∫—Å –∏ –∑–Ω–∞—á–µ–Ω–∏–µ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ

**–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ:**
- –í –ø—Ä–∏–º–µ—Ä–µ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è `enumerate(todo_list, 1)` ‚Äî –Ω—É–º–µ—Ä–∞—Ü–∏—è –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è —Å 1

````python
# –°–ø–∏—Å–æ–∫ –¥–µ–ª –§—É–Ω—Ç–∏–∫–∞ üìã

todo_list = [
    "üìö –ù–∞–ø–∏—Å–∞—Ç—å —ç—Å—Å–µ",
    "üíª –ó–∞–¥–∞–Ω–∏–µ –ø–æ Python",
    "üé¨ –õ–µ–∫—Ü–∏—è –Ω–∞ YouTube",
    "‚òï –ö–æ—Ñ–µ —Å –¥—Ä—É–∑—å—è–º–∏",
    "üéÆ –í–∏–¥–µ–æ–∏–≥—Ä—ã"
]

# –í—ã–≤–µ—Å—Ç–∏ –≤–µ—Å—å —Å–ø–∏—Å–æ–∫
print("üê± –î–µ–ª–∞ –§—É–Ω—Ç–∏–∫–∞:")
print(todo_list)

# –î–æ—Å—Ç—É–ø –ø–æ –∏–Ω–¥–µ–∫—Å—É
print("\n–°–∞–º–æ–µ –≤–∞–∂–Ω–æ–µ:")
print(todo_list[0])

print("\n–°–∞–º–æ–µ –ø—Ä–∏—è—Ç–Ω–æ–µ:")
print(todo_list[-1])

# –î–æ–±–∞–≤–∏—Ç—å –Ω–æ–≤–æ–µ –¥–µ–ª–æ
todo_list.append("üèãÔ∏è –°–ø–æ—Ä—Ç–∑–∞–ª")

# –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–µ–ª
count = len(todo_list)
print(f"\nüê± –í—Å–µ–≥–æ –¥–µ–ª: {count}")

# –¶–∏–∫–ª –ø–æ —Å–ø–∏—Å–∫—É (–ø–µ—Ä–≤—ã–µ 3)
print("\n–ü–µ—Ä–≤—ã–µ 3 –¥–µ–ª–∞:")
for i in range(3):
    print(f"  {i+1}. {todo_list[i]}")
````

````python
# –ë–∞–π—Ç–∏–∫ –ø—Ä–æ–≤–µ—Ä—è–µ—Ç –∑–∞–≤—Ç—Ä–∞–∫ ü•Ñ
mice = 3  # –º—ã—à–∫–∏ –Ω–∞ –∑–∞–≤—Ç—Ä–∞–∫
sausages = 1  # —Å–æ—Å–∏—Å–∫–∏

if mice >= 3:
    print("‚úÖ –û—Ç–ª–∏—á–Ω—ã–π –∑–∞–≤—Ç—Ä–∞–∫!")
elif mice >= 1 and sausages > 0:
    print("‚ö†Ô∏è –ù–æ—Ä–º–∞–ª—å–Ω—ã–π –∑–∞–≤—Ç—Ä–∞–∫")
else:
    print("‚ùå –ì–¥–µ –º–æ–π –∑–∞–≤—Ç—Ä–∞–∫?!")
    print("–Ø —Ç—Ä–µ–±—É—é –∑–∞–≤—Ç—Ä–∞–∫!")
````

## üê± –§—É–Ω–∫—Ü–∏–∏ ‚Äî –º–Ω–æ–≥–æ–∫—Ä–∞—Ç–Ω–æ–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –∫–æ–¥–∞

–§—É–Ω–∫—Ü–∏—è ‚Äî —ç—Ç–æ –±–ª–æ–∫ –∫–æ–¥–∞ —Å –∏–º–µ–Ω–µ–º, –∫–æ—Ç–æ—Ä—ã–π –º–æ–∂–Ω–æ –≤—ã–∑—ã–≤–∞—Ç—å –º–Ω–æ–≥–æ–∫—Ä–∞—Ç–Ω–æ.

**–ó–∞—á–µ–º –Ω—É–∂–Ω—ã —Ñ—É–Ω–∫—Ü–∏–∏?**
- –ù–µ –ø–æ–≤—Ç–æ—Ä—è—Ç—å –æ–¥–∏–Ω –∏ —Ç–æ—Ç –∂–µ –∫–æ–¥
- –°–¥–µ–ª–∞—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º—É –ø–æ–Ω—è—Ç–Ω–µ–µ
- –õ–µ–≥–∫–æ –≤–Ω–æ—Å–∏—Ç—å –∏–∑–º–µ–Ω–µ–Ω–∏—è

**–°—Ç—Ä—É–∫—Ç—É—Ä–∞:**
```python
def –∏–º—è_—Ñ—É–Ω–∫—Ü–∏–∏():
    # –∫–æ–¥ —Ñ—É–Ω–∫—Ü–∏–∏
    print("–ß—Ç–æ-—Ç–æ –¥–µ–ª–∞–µ–º")

# –í—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏
–∏–º—è_—Ñ—É–Ω–∫—Ü–∏–∏()
```

**–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ:** –≤–º–µ—Å—Ç–æ 6 —Å—Ç—Ä–æ–∫ `print` –º—ã –ø–∏—à–µ–º –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É `introduce()`!

````python
# –°–æ–∑–¥–∞—ë–º —Ñ—É–Ω–∫—Ü–∏—é –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è
def introduce():
    print("="*50)
    print("üê± –ü—Ä–∏–≤–µ—Ç, —è –∫–æ—Ç –ë–∞–π—Ç–∏–∫!")
    print("üìç –°—Ç—É–¥–µ–Ω—Ç 2 –∫—É—Ä—Å–∞")
    print("üè´ –ò–∑—É—á–∞—é Data Science")
    print("üíª –Ø–∑—ã–∫: Python")
    print("="*50)

# –í—ã–∑—ã–≤–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é –æ–¥–∏–Ω —Ä–∞–∑
introduce()
````

- `total_food` –æ–±—ä—è–≤–ª–µ–Ω–∞ –≤–Ω–µ —Ñ—É–Ω–∫—Ü–∏–∏ (–≥–ª–æ–±–∞–ª—å–Ω–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è)
- –í–Ω—É—Ç—Ä–∏ —Ñ—É–Ω–∫—Ü–∏–∏ –∏—Å–ø–æ–ª—å–∑—É–µ–º `global total_food` –¥–ª—è –µ—ë –∏–∑–º–µ–Ω–µ–Ω–∏—è
- –ö–∞–∂–¥—ã–π –≤—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–µ—Ä–µ–¥–∞—ë—Ç —Ä–∞–∑–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ: 50, 70, 60

````python
# –§—É–Ω—Ç–∏–∫ —Å—á–∏—Ç–∞–µ—Ç, —Å–∫–æ–ª—å–∫–æ –æ–Ω —Å—ä–µ–ª –∑–∞ –¥–µ–Ω—å üçΩÔ∏è

total_food = 0  # –ü–µ—Ä–µ–º–µ–Ω–Ω–∞—è –¥–ª—è –Ω–∞–∫–æ–ø–ª–µ–Ω–∏—è –µ–¥—ã

def add_food(amount):
    global total_food
    total_food = total_food + amount
    print(f"üê± –§—É–Ω—Ç–∏–∫ —Å—ä–µ–ª {amount} –≥. –í—Å–µ–≥–æ: {total_food} –≥")

# –ó–∞–≤—Ç—Ä–∞–∫
add_food(50)

# –û–±–µ–¥
add_food(70)

# –£–∂–∏–Ω
add_food(60)

# –§–∏–Ω–∞–ª—å–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç
print(f"\nüìä –ò—Ç–æ–≥–æ –∑–∞ –¥–µ–Ω—å: {total_food} –≥—Ä–∞–º–º–æ–≤")
````

## üí∞ –§—É–Ω–∫—Ü–∏–∏ —Å –Ω–µ—Å–∫–æ–ª—å–∫–∏–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏

–§—É–Ω–∫—Ü–∏—è –º–æ–∂–µ—Ç –ø—Ä–∏–Ω–∏–º–∞—Ç—å –Ω–µ –æ–¥–∏–Ω, –∞ –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤.

**–í —ç—Ç–æ–º –ø—Ä–∏–º–µ—Ä–µ:**
- –î–≤–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞: `amount` (—Å—É–º–º–∞) –∏ `name` (–æ–ø–∏—Å–∞–Ω–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–∏)
- –£—Å–ª–æ–≤–∏–µ –≤–Ω—É—Ç—Ä–∏ —Ñ—É–Ω–∫—Ü–∏–∏: –ø—Ä–æ–≤–µ—Ä—è–µ–º, –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–∞—è –∏–ª–∏ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–∞—è —Å—É–º–º–∞
- –ò—Å–ø–æ–ª—å–∑—É–µ–º —Ä–∞–∑–Ω—ã–µ —ç–º–æ–¥–∑–∏ –¥–ª—è –¥–æ—Ö–æ–¥–æ–≤ (‚úÖ) –∏ —Ä–∞—Å—Ö–æ–¥–æ–≤ (‚ùå)

**–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ:** —Ñ—É–Ω–∫—Ü–∏—è —Å–∞–º–∞ —Ä–µ—à–∞–µ—Ç, –∫–∞–∫ —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞—Ç—å –≤—ã–≤–æ–¥ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –∑–Ω–∞–∫–∞ —á–∏—Å–ª–∞!

````python
# –ë–∞–ª–∞–Ω—Å –ë–∞–π—Ç–∏–∫–∞ –≤ –±–∏—Ç–∫–æ–π–Ω–∞—Ö ‚Çø

balance = 0

def transaction(amount, name):
    global balance
    balance = balance + amount

    if amount > 0:
        print(f"‚úÖ {name}: +{amount}‚Çø")
    else:
        print(f"‚ùå {name}: {amount}‚Çø")

    print(f"   –ë–∞–ª–∞–Ω—Å: {balance}‚Çø\n")

# –¢—Ä–∞–Ω–∑–∞–∫—Ü–∏–∏
transaction(100, "–ó–∞—Ä–ø–ª–∞—Ç–∞")
transaction(-30, "–ö–æ—Ä–º")
transaction(-20, "–ò–≥—Ä—É—à–∫–∏")
transaction(50, "–ü—Ä–µ–º–∏—è")

print(f"üí∞ –ò—Ç–æ–≥–æ: {balance}‚Çø")
````
