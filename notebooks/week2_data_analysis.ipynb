## üì• [2] Load & Prepare Data

–ó–∞–≥—Ä—É–∂–∞–µ–º CSV-—Ñ–∞–π–ª—ã –∏–∑ –ø–∞–ø–∫–∏ `data/examples/` –∏ –≤—ã–ø–æ–ª–Ω—è–µ–º –±–∞–∑–æ–≤—É—é –ø–æ–¥–≥–æ—Ç–æ–≤–∫—É:
- –£–¥–∞–ª—è–µ–º —Å—Ç–æ–ª–±—Ü—ã —Å URL Wikidata
- –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞–µ–º —Å—Ç–æ–ª–±—Ü—ã `filmLabel` ‚Üí `film`, `genreLabel` ‚Üí `genre`, `countryLabel` ‚Üí `country`
- –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ç–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö

```python
import pandas as pd
import numpy as np

# === PATHS TO CSV FILES ===
data_dir = 'data/examples'
file1 = 'cartoons_capital_cost.csv'
file2 = 'cartoons_genre_country_duration.csv'

path1 = os.path.join(data_dir, file1)
path2 = os.path.join(data_dir, file2)

print(f"üìÇ Loading data from:")
print(f"  1. {path1}")
print(f"  2. {path2}\n")

# === LOAD CSV FILES ===
try:
    df_cost = pd.read_csv(path1)
    print(f"‚úÖ Loaded {file1}: {len(df_cost)} rows")

    df_genre = pd.read_csv(path2)
    print(f"‚úÖ Loaded {file2}: {len(df_genre)} rows\n")

except FileNotFoundError as e:
    print(f"‚ùå Error: {e}")
    print("–£–±–µ–¥–∏—Ç–µ—Å—å, —á—Ç–æ CSV-—Ñ–∞–π–ª—ã –Ω–∞—Ö–æ–¥—è—Ç—Å—è –≤ –ø–∞–ø–∫–µ 'data/examples/'")
    raise

# === DATA CLEANING & RENAMING ===

# 1Ô∏è‚É£ Process cartoons_capital_cost.csv
print("üîß Processing cartoons_capital_cost.csv...")

# –£–¥–∞–ª—è–µ–º —Å—Ç–æ–ª–±–µ—Ü —Å URL Wikidata (–ø–µ—Ä–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü 'film')
if 'film' in df_cost.columns and df_cost['film'].str.startswith('http://www.wikidata.org').any():
    df_cost = df_cost.drop(columns=['film'])
    print("  ‚ûñ Removed Wikidata URL column")

# –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞–µ–º filmLabel ‚Üí film
if 'filmLabel' in df_cost.columns:
    df_cost = df_cost.rename(columns={'filmLabel': 'film'})
    print("  ‚úèÔ∏è  Renamed 'filmLabel' ‚Üí 'film'")

# –ü—Ä–æ–≤–µ—Ä—è–µ–º —á–∏—Å–ª–æ–≤–æ–π —Ç–∏–ø –¥–ª—è capital_cost
df_cost['capital_cost'] = pd.to_numeric(df_cost['capital_cost'], errors='coerce').fillna(0).astype(int)
print(f"  ‚úÖ Converted 'capital_cost' to numeric (int)\n")


# 2Ô∏è‚É£ Process cartoons_genre_country_duration.csv
print("üîß Processing cartoons_genre_country_duration.csv...")

# –£–¥–∞–ª—è–µ–º —Å—Ç–æ–ª–±–µ—Ü —Å URL Wikidata
if 'film' in df_genre.columns and df_genre['film'].str.startswith('http://www.wikidata.org').any():
    df_genre = df_genre.drop(columns=['film'])
    print("  ‚ûñ Removed Wikidata URL column")

# –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞–µ–º —Å—Ç–æ–ª–±—Ü—ã
rename_map = {
    'filmLabel': 'film',
    'genreLabel': 'genre',
    'countryLabel': 'country'
}
df_genre = df_genre.rename(columns=rename_map)
print("  ‚úèÔ∏è  Renamed columns: filmLabel‚Üífilm, genreLabel‚Üígenre, countryLabel‚Üícountry")

# –ü—Ä–æ–≤–µ—Ä—è–µ–º —á–∏—Å–ª–æ–≤–æ–π —Ç–∏–ø –¥–ª—è duration
df_genre['duration'] = pd.to_numeric(df_genre['duration'], errors='coerce').fillna(0).astype(int)
print(f"  ‚úÖ Converted 'duration' to numeric (int)\n")

print("=" * 60)
print("‚úÖ Data loading and cleaning complete!")
print("=" * 60)
```
